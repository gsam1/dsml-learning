```python
preset = 'file:///root/lab/ws/dsml-learning/spark_python_handson/dataset/'
```

```python
import re
from pyspark import SparkConf, SparkContext

conf = SparkConf().setAppName('WordsCount')
sc = SparkContext(conf = conf)
```

```python
def normalizeWords(text):
    return re.compile(r'\W+', re.UNICODE).split(text.lower())

lines = sc.textFile(preset + 'Book.txt')
words = lines.flatMap(normalizeWords)
wordCounts = words.countByValue()

for word, count in wordCounts.items():
    cleanWord = word.encode('ascii', 'ignore')
    if (cleanWord):
        print(cleanWord.decode() + " " + str(count))
```

```python
newWordCounts = words.map(lambda x: (x,1)).reduceByKey(lambda x, y: x + y)
sortedWordCounts = newWordCounts.map(lambda x: (x[1],x[0])).sortByKey(ascending=False)
```

```python
results = sortedWordCounts.collect()

for result in results:
    count = result[0]
    word = result[1].encode('ascii', 'ignore')
    if (word):
        print(f'{word}:\t\t{count}')
```
